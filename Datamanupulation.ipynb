Running multiple threads (smaller units of a process) within the same process. Threads share the same memory space.
   - Multiprocessing -
Running multiple processes (independent programs) simultaneously, each with its own memory space.

2.   Challenges Associated with Memory Management in Python -
     
  * Memory Leaks (especially with Circular References):
While Python's garbage collector handles most memory deallocation through reference counting, circular references can prevent objects from being collected. If two or more objects reference each other, their reference counts may never drop to zero, even if they are no longer accessible by the program, leading to memory leaks.

  * High Memory Usage and Bloat:
Python objects often have a higher memory footprint compared to equivalent data in lower-level languages like C. The interpreter also holds onto freed memory within its private heap rather than immediately returning it to the operating system, potentially leading to higher overall memory usage, especially in long-running applications.

   * Performance Overhead of Garbage Collection:
The garbage collection process, while automatic, can introduce performance overhead. When the collector runs, it can pause program execution to identify and reclaim unreferenced objects, which can be noticeable in performance-critical applications.
   
   * Limited Manual Control:
Python abstracts away much of the low-level memory management, offering less direct control compared to languages where manual memory allocation and deallocation are common. This can make fine-tuning memory usage for specific performance optimizations more challenging.

  * Fragmentation:
Memory fragmentation can occur within Python's private heap, where allocated and deallocated blocks leave small, unusable gaps. This can hinder the allocation of larger contiguous memory blocks, even if sufficient total memory is available.


```python
# 3 :Write a Python program that logs an error message to a log file when a division by zero exception occurs.

import logging

logging.basicConfig(filename="error_log.txt",
                    level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

def divide(a, b):
    try:
        result = a / b
        print("Result:", result)
    except ZeroDivisionError as e:
        logging.error("Division by zero error occurred: %s", e)
        print("Error: Division by zero is not allowed!")

divide(10, 2)
divide(5, 0)

```

```python
# 4: Write a Python program that reads from one file and writes its content to another file.

# Create a sample source file
with open("source.txt", "w") as f:
    f.write("Hello, this is the source file.\n")
    f.write("Python file handling in Google Colab is easy!\n")
    f.write("We are copying this content to another file.")

# Read from source file and write to destination file
with open("source.txt", "r") as src_file:
    content = src_file.read()

with open("destination.txt", "w") as dest_file:
    dest_file.write(content)

# Display the content of both files
print(" Content of source.txt:")
with open("source.txt", "r") as f:
    print(f.read())

print("\n Content of destination.txt:")
with open("destination.txt", "r") as f:
    print(f.read())


```

```python
# 5: Write a program that handles both IndexError and KeyError using a try-except block.

try:
    my_list = [10, 20, 30]
    print("Accessing list element:", my_list[5])

    my_dict = {"name": "Ishu", "age": 25}
    print("Accessing dictionary key:", my_dict["city"])

except IndexError:
    print("IndexError: You tried to access an index that doesn’t exist in the list.")

except KeyError:
    print("KeyError: You tried to access a key that doesn’t exist in the dictionary.")

finally:
    print("Program executed successfully (with or without errors).")


```

6.   What are the differences between NumPy arrays and Python lists?

  - NumPy Arrays:
  
     * Must contain elements of a single, homogeneous data type (e.g., all integers, all floats). This allows for efficient memory allocation and optimized mathematical operations.
     * Offer significantly better performance for numerical computations, especially with large datasets, due to their fixed size, homogeneous data type, and underlying C/Fortran implementations. They leverage contiguous memory allocation, which improves data access and processing speed.

   - Python Lists:
   
     * Can contain elements of heterogeneous data types (e.g., integers, strings, and other objects within the same list).
     * Are generally slower for numerical operations, particularly on large datasets, as they store references to objects and require more overhead for type checking and dynamic resizing.

7.   Explain the difference between apply() and map() in Pandas.

  - map():
     
      * Used only on Pandas Series (single column).
      * It applies a function element-wise to each value in a Series.
      * It is best for simple operations on individual elements.
      * It can take a function, dictionary, or Series for element-wise mapping.
      * It is usually faster because it works on one column.

  - apply():
  
      * it Can be used on both Series and DataFrames.
      * It can apply a function across rows or columns (axis=0 or axis=1) in a DataFrame.
      * It is used for complex operations that may involve multiple columns or rows.
      * it is generally takes a function (custom, built-in, or lambda).
      * It can be slower when used on DataFrames due to more complex operations.

```python
# 10: Calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Maths': [78, 85, 96, 80, 86],
    'Science': [84, 95, 92, 75, 90],
    'English': [70, 88, 82, 65, 78],
    'History': [68, 80, 75, 60, 72]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

print("✅ Correlation Matrix:\n")
print(corr_matrix)

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
